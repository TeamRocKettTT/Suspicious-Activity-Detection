```python
base_dir = 'C:/Users/keval/Desktop/Capstone project/V3 Tiny/TrainVideos'
classes = ['Accident', 'Fire', 'Burglary', 'Fighting', 'Normal', 'Shooting']

def extract_frames(video_path, target_dir, frames_per_second=1):
    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    frames = []
    frame_rate = cap.get(cv2.CAP_PROP_FPS)
    frame_interval = int(round(frame_rate / frames_per_second))
    
    while cap.isOpened():
        ret, frame = cap.read()
        if ret:
            if frame_count % frame_interval == 0:
                frame_path = os.path.join(target_dir, f"{frame_count}.jpg")
                cv2.imwrite(frame_path, frame)
                frames.append(frame_path)
            frame_count += 1
        else:
            break
    
    cap.release()
    return frames

for category in classes:
    category_dir = os.path.join(base_dir, category)
    frames_dir = os.path.join(category_dir, 'frames')
    os.makedirs(frames_dir, exist_ok=True)
    
    for video_file in os.listdir(category_dir):
        video_path = os.path.join(category_dir, video_file)
        if video_file.endswith('.mp4'):
            extract_frames(video_path, frames_dir, frames_per_second=1)  # Extract frames at 1 frame per second
```

# Model Building

```python
def YOLOv3(input_shape, num_classes):
    # Define YOLO model architecture
    inputs = tf.keras.Input(shape=input_shape)
    
    # Backbone
    x = layers.Conv2D(32, (3, 3), strides=(1, 1), padding='same', activation='relu')(inputs)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Conv2D(64, (3, 3), strides=(1, 1), padding='same', activation='relu')(x)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Conv2D(128, (3, 3), strides=(1, 1), padding='same', activation='relu')(x)
    x = layers.MaxPooling2D((2, 2))(x)
    
    # Detection head
    x = layers.Conv2D(256, (3, 3), strides=(1, 1), padding='same', activation='relu')(x)
    x = layers.Conv2D(128, (1, 1), strides=(1, 1), padding='same', activation='relu')(x)
    x = layers.Conv2D(256, (3, 3), strides=(1, 1), padding='same', activation='relu')(x)
    x = layers.Conv2D(128, (1, 1), strides=(1, 1), padding='same', activation='relu')(x)
    x = layers.Conv2D(256, (3, 3), strides=(1, 1), padding='same', activation='relu')(x)
    
    # Output
    outputs = layers.Conv2D(num_classes, (1, 1), strides=(1, 1), padding='same', activation='softmax')(x)
    
    model = tf.keras.Model(inputs, outputs)
    return model
```

```python
def resize_frame(frame_path, target_shape):
    # Load frame using OpenCV
    frame = cv2.imread(frame_path)
    if frame is not None:
        # Resize frame to target shape
        resized_frame = cv2.resize(frame, target_shape)
        return resized_frame
    else:
        return None

def resize_frames_in_directory(data_dir, target_shape):
    # Create a new directory to save resized frames
    output_dir = os.path.join(data_dir, "resized_frames")
    os.makedirs(output_dir, exist_ok=True)

    # Iterate over all files in the data directory
    for file_name in sorted(os.listdir(data_dir)):
        if file_name.endswith('.jpg') or file_name.endswith('.png'):
            frame_path = os.path.join(data_dir, file_name)
            resized_frame = resize_frame(frame_path, target_shape)
            if resized_frame is not None:
                # Save resized frame to output directory
                output_path = os.path.join(output_dir, file_name)
                cv2.imwrite(output_path, resized_frame)
                print(f"Resized frame saved: {output_path}")
            else:
                print(f"Unable to resize frame: {frame_path}")

# Usage example
data_dir = 'C:/Users/keval/Desktop/Capstone project/V3 Tiny/TrainVideos/VideosT'
target_shape = (360, 360)  # Define your target shape (height, width)
resize_frames_in_directory(data_dir, target_shape)
```

```python
def load_data(data_dir, target_shape):
    frames = []
    annotations = []

    # Iterate over all files in the data directory
    for file_name in sorted(os.listdir(data_dir)):
        if file_name.endswith('.jpg') or file_name.endswith('.png'):
            frame_path = os.path.join(data_dir, file_name)
            annotation_path = os.path.join(data_dir, os.path.splitext(file_name)[0] + '.txt')

            # Check if annotation file exists
            if os.path.exists(annotation_path):
                # Load frame using OpenCV
                frame = cv2.imread(frame_path)

                # Check if frame is None (e.g., unable to read frame)
                if frame is None:
                    print(f"Unable to read frame: {frame_path}")
                    continue

                # Resize frame to target shape
                frame = cv2.resize(frame, target_shape)

                # Read annotation
                with open(annotation_path, 'r') as f:
                    annotation_lines = f.readlines()

                # Process each annotation line
                for annotation_line in annotation_lines:
                    annotation = annotation_line.strip().split()

                    # Extract class index and bounding box coordinates
                    class_index = int(annotation[0])
                    bbox = list(map(float, annotation[1:]))

                    # Append frame and annotation to lists
                    frames.append(frame)
                    annotations.append([class_index] + bbox)

            else:
                print(f"Annotation file not found for frame: {frame_path}. Skipping frame.")

    # Convert lists to NumPy arrays
    frames = np.array(frames)
    annotations = np.array(annotations)

    return frames, annotations
```

```python
def yolo_loss(y_true, y_pred):
    # Define the number of bounding box coordinates and classes
    num_boxes = 4  # Assuming each bounding box has 4 coordinates (x, y, width, height)
    num_classes = 5  # Update this based on your dataset
    
    # Compute the number of channels in the predictions
    num_channels = num_boxes + 1 + num_classes  # x, y, width, height, confidence, class probabilities
    
    # Split predictions and ground truth tensors
    pred_xy, pred_wh, pred_confidence, pred_class_probs = tf.split(y_pred, (2, 2, 1, num_classes), axis=-1)
    true_xy, true_wh, true_confidence, true_class_probs = tf.split(y_true, (2, 2, 1, num_classes), axis=-1)
    
    # Compute object mask: 1 if object present, else 0
    object_mask = true_confidence
    
    # Compute no object mask: 1 if no object present, else 0
    no_object_mask = 1 - object_mask
    
    # Compute confidence loss (objectness loss)
    confidence_loss = object_mask * tf.square(pred_confidence - true_confidence) + \
                      0.5 * no_object_mask * tf.square(pred_confidence - true_confidence)
    
    # Compute localization loss
    localization_loss = object_mask * (tf.square(pred_xy - true_xy) + \
                                        tf.square(pred_wh - true_wh))
    
    # Compute classification loss
    classification_loss = object_mask * tf.square(pred_class_probs - true_class_probs)
    
    # Sum all the individual losses
    total_loss = tf.reduce_sum(confidence_loss + localization_loss + classification_loss)
    
    return total_loss
```

```python
train_datagen = ImageDataGenerator(rescale=1./255, 
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   rotation_range=40,
                                   zoom_range=0.2,
                                   horizontal_flip=True)

val_datagen = ImageDataGenerator(rescale=1./255)

# Flow training images and annotations in batches of 32 using train_datagen generator
train_generator = train_datagen.flow(train_frames, train_annotations, batch_size=32)

# Flow validation images and annotations in batches of 32 using val_datagen generator
val_generator = val_datagen.flow(val_frames, val_annotations, batch_size=32)

# Compile and Train Model
input_shape = (360, 360, 3)  # Add the channel dimension
num_classes = 5
model = YOLOv3(input_shape=input_shape, num_classes=num_classes)
model.compile(optimizer='adam', loss=yolo_loss, metrics=['accuracy'])

# Calculate steps_per_epoch and validation_steps
steps_per_epoch = len(train_frames) // 32
validation_steps = len(val_frames) // 32

# Train the model using the generators
model.fit(train_generator, epochs=10, steps_per_epoch=steps_per_epoch, 
          validation_data=val_generator, validation_steps=validation_steps)
```
